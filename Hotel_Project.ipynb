## Data Visualization

```python
df['adr'].plot(kind ='box')
```

```python
df = df[df['adr']<5000]
```

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not cancled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
city_hotel = df[df['hotel'] == 'City Hotel']
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in City and Resort Hotels', fontsize=30)

plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', linestyle='-', marker='o', color='royalblue', linewidth=2)
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel', linestyle='--', marker='s', color='tomato', linewidth=2)

plt.legend(fontsize=20)
plt.xlabel('Reservation Status Date', fontsize=20)
plt.ylabel('Average Daily Rate (ADR)', fontsize=20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16, 8))

ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')

legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))

plt.title('Reservation Status per Month', size=20)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Number of Reservations', fontsize=15)

plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

```python
adr_per_month_canceled = df[df['is_canceled'] == 1].groupby('month')['adr'].mean().reset_index()
print(adr_per_month_canceled.head())
```

```python
plt.figure(figsize=(15, 8))

sns.barplot(x='month', y='adr', data=adr_per_month_canceled, palette='coolwarm')

plt.title('ADR per Month (Canceled Reservations)', fontsize=20)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Total ADR', fontsize=15)

plt.legend(['ADR for Canceled Reservations'], fontsize=20)
plt.show()

```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 Nations with reservations cancelled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
# Define the DataFrame for canceled and not canceled reservations
cancelled_data = df[df['is_canceled'] == 1]   
not_cancelled_df = df[df['is_canceled'] == 0] 

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_df_adr = not_cancelled_df.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)
plt.title('Average Daily Rate (ADR) Comparison', fontsize=20)

plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.legend(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate (ADR) Comparison', fontsize=20)

plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')

plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')

plt.legend(fontsize=15)
plt.show()
```

# Filter the data for dates from 2016 onwards

```python
cancelled_df_adr_filtered = cancelled_df_adr[cancelled_df_adr['reservation_status_date'] >= '2016-01-01']
not_cancelled_df_adr_filtered = not_cancelled_df_adr[not_cancelled_df_adr['reservation_status_date'] >= '2016-01-01']

plt.figure(figsize=(20, 6))

plt.title('Average Daily Rate (ADR) Comparison (2016 Onwards)', fontsize=25)

plt.plot(not_cancelled_df_adr_filtered['reservation_status_date'], not_cancelled_df_adr_filtered['adr'], label='Not Cancelled')

plt.plot(cancelled_df_adr_filtered['reservation_status_date'], cancelled_df_adr_filtered['adr'], label='Cancelled')

plt.legend(fontsize=25)
plt.show()
```
